```python
import pandas as pd

def read_year_csv(folder_path, year):
    file_path = f"{folder_path}/wheat-coords-{year}.csv"
    df = pd.read_csv(file_path)
    return df
```

```python
import requests
import json

years = [2017, 2018, 2019, 2020, 2021, 2022, 2023]

for year in years:
    df = read_year_csv('/Users/ben/dev/projects/LateBloomers/test stuff/wheat-coords', year)

    data = []
    for entry in df.itertuples():
        payload = {
            "latitude": entry.latitude,
            "longitude": entry.longitude,
            "year": year
        }
        print("Requesting:", payload)
        try:
            r = requests.post("http://127.0.0.1:5000/getClimateData", json=payload, timeout=20)
        except requests.Timeout:
            print("Request timed out for:", payload)
            continue
        r.raise_for_status()
        if r.json():
            data.append(r.json())

    with open(f'wheat-data/{year}.json', 'w') as f:
        json.dump(data, f, indent=2)
```
